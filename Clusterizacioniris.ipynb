```python
# Paso 1: Importar bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Paso 2: Cargar el conjunto de datos
iris = load_iris()
X = iris.data
y = iris.target
```

```python
#Aplicar el método del codo
inertia_values = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

# Visualización del método del codo
plt.plot(k_values, inertia_values, marker='o')
plt.xlabel('Número de clusters')
plt.ylabel('Inercia')
plt.title('Método del Codo')
plt.show()
```

```python
k_optimo = 3  # Número óptimo de clusters determinado por el método del codo
kmeans = KMeans(n_clusters=k_optimo, random_state=42)
kmeans.fit(X)
y_pred = kmeans.labels_
```

```python
# Comparar los clusters generados por K-means con las etiquetas originales del conjunto de datos Iris
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k', s=20)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Etiquetas originales')

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', edgecolor='k', s=20)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', c='red', s=100, label='Centroides')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clusters generados por K-means')
plt.legend()
plt.show()
```
